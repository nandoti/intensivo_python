### Aula02 - Automatização de Processos

- Passo 1: buscar os dados das ações automaticamente
- passo 2: realizar as análises solicitadas pelo gestor
- passo 3: enviar automatuicamente o email com o resultado da análise

### Passo 1

```python
import yfinance as yf

ticker = input("Digite o código da ação: ")
dados = yf.Ticker(ticker).history("6mo")
fechamento = dados.Close
fechamento.plot()
```

### Passo 2 - realizar as análises solicitadas pelo gestor

- cotação máxima
- cotação mínima
- cotação atual

```python
maxima = round(fechamento.max(),2)
minima = round(fechamento.min(),2)
cotacao_atual = round(fechamento[-1],2) # pega o índice desejado. Pode ser usado índice negativo

display(maxima)
display(minima)
display(cotacao_atual)
```

### Passo 3: enviar automatuicamente o email com o resultado da análise

```python
import time

import pyautogui
import pyperclip
```

```python
pyautogui.PAUSE = 2

pyautogui.hotkey("ctrl", "t")
pyperclip.copy("www.gmail.com")
pyautogui.hotkey("ctrl","v")
pyautogui.hotkey("enter")

pyautogui.click(x=110, y=209)
pyperclip.copy("nando.frontend@gmail.com")
pyautogui.hotkey("ctrl","v")
pyautogui.hotkey("tab")

pyperclip.copy("Análises diárias")
pyautogui.hotkey("ctrl","v")
pyautogui.hotkey("tab")

mensagem = f"""
Prezado gestor,

Seguem, conforme solicitado, as análises dos últimos seis meses da ação {ticker}:

Cotação máxima: {maxima}
Cotação mínima: {minima}
Cotação Atual: {cotacao_atual}

Qualquer dúvida, fico à disposição!

Atte.
Fernando

"""
pyperclip.copy(mensagem)
pyautogui.hotkey("ctrl","v")

pyautogui.click(x=622, y=1000)
display("Email enviado com sucesso!")
```
